```python
# Таким способом можно посчитать значение синуса любого угла, но видимо требовалось не такое решение задачи :)
import math

x = float(input("Введите значение угла для рассчета синуса [0, 30]: "))

#print(math.sin(x))

sinus = math.sin(math.radians(x))

print(f'sin(x) = {sinus}')
```
