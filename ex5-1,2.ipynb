```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

from pylab import *
from mpl_toolkits.mplot3d import Axes3D

fig = figure()
ax = Axes3D(fig)

x = np.arange(-10, 10, 0.1)
y = np.arange(-10, 10, 0.5)
x, y = np.meshgrid(x, y)
z = 20 * x + 30 * y
z1 = 20 *x + 30 * y - 400
ax.plot_surface(x, y, z)
ax.plot_wireframe(x, y, z1)

show()
```
