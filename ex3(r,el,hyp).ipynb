```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# ROUND
x = []
y1 = []
y2 = []
r = 500
for i in range(-500,501):
    x_ = i
    x.append(x_)
    y1.append(math.sqrt(r ** 2 - x_**2))
    y2.append(-math.sqrt(r ** 2 - x_**2))
plt.plot(x, y1,500)
plt.plot(x, y2,500)
plt.xlabel('x      ROUND')
plt.ylabel('y')
```

```python
# ELLIPS
x = []
y1 = []
y2 = []
a = 200
b = 300
for i in range(-200,201):
    x_ = i
    x.append(x_)
    y1.append(math.sqrt(((a**2) * (b**2) - (x_**2) * (b**2))/a**2))
    y2.append(-math.sqrt(((a**2) * (b**2) - (x_**2) * (b**2))/a**2))

plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x     ELLIPSE')
plt.ylabel('y')
```

```python
# HYP
x1 = []
x2 = []

y1 = []
y2 = []
y3 = []
y4 = []
a = 200
b = 300
for i in range(-1800,-199):
    x_ = i
    x1.append(x_)
    y1.append(math.sqrt((((x_**2) * (b**2))/(a**2) - b**2)))
    y2.append(-math.sqrt((((x_**2) * (b**2))/(a**2) - b**2)))

for i in range(200,1800):
    x_ = i
    x2.append(x_)
    y3.append(math.sqrt((((x_**2) * (b**2))/(a**2) - b**2)))
    y4.append(-math.sqrt((((x_**2) * (b**2))/(a**2) - b**2)))

plt.plot(x1, y1)
plt.plot(x1, y2)
plt.plot(x2, y3)
plt.plot(x2, y4)

plt.xlabel('x     HYP')
plt.ylabel('y')
```
