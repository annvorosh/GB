```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-4*np.pi, 4*np.pi, 201)
y1 = 1 * cos(x - 0) + 1
y2 = 2 * cos(x - 0) + 1
y3 = 1 * cos(x - 2*np.pi/2) + 1
y4 = 1 * cos(x - 0) + 2
plt.plot(x, y1, 'red')
plt.plot(x, y2, 'green')
plt.plot(x, y3, 'yellow')
plt.plot(x, y4, 'purple')
plt.xlabel('Ось Х')
plt.ylabel('Ось Y')
plt.grid(True)
plt.show()
```
