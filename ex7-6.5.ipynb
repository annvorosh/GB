```python
#     :
# x + 2y  z = 1  
# 8x  5y + 2z = 12
#     Q(x,y,z),   ,    . 


%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from pylab import *

fig = figure()
ax = Axes3D(fig)

def Q(x, y, z):
    return (x**2 + y**2 + z**2)
x = np.arange(-5, 7, 0.3)
y = np.arange(-5, 5, 0.3)
x, y = np.meshgrid(x, y)

#  x + 2y -  z = 1   \ *2   ->   2x + 4y - 2z = 2     ->  10x - y = 14
# 8x - 5y + 2z = 12              8x - 5y + 2z = 12        y = 10x - 14

#  x + 2y -  z = 1   \ *5   ->    5x + 10y - 5z = 5     ->  21x - z = 29
# 8x - 5y + 2z = 12  \ *2        16x - 10y + 4z = 24        z = 21x - 29

ax.plot_surface(x, y, Q(x, 10 * x - 14, 21 * x - 29))
ax.scatter(xs = 1.38191882, ys = -0.18081181, zs = 0.0202952, c = 'red')
show()

fig = figure()
ax = Axes3D(fig)
x = np.arange(-5, 7, 0.3)
y = np.arange(-5, 5, 0.3)
y = np.arange(-3, 100, 0.3)
x, y = np.meshgrid(x, y)
z = x + 2 * y - 1
z1 = 6 - 4 * x + 2.5 * y
ax.plot_surface(y, x, z)
ax.plot_surface(y, x, z1)
ax.scatter(xs = 1.38191882, ys = -0.18081181, zs = 0.0202952, c = 'black')
show()
```
