```python
# код переводит полярные координаты в декартовы

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
R = np.linspace(0, 10, 3)
a = np.linspace(0, 2*np.pi, 3)
x = R * np.cos(a)
print(f'Даны полярные координаты: \nрадиус {R}, \nугол {a}')
print(f'Им соответствуют значения x: {x}')
y = R * np.sin(a)
print(f'И значения y: {y}')
```

```python
# КОД РИСУЕТ ГРАФИК ОКРУЖНОСТИ В ПОЛЯРНЫХ КООРДИНАТАХ

a1 = np.linspace(0, 2 * np.pi, 100)
R1 = 0
b1 = R1*np.cos(a1) + 3
plt.polar(a1, b1, 'purple')
plt.xlabel('ROUND')
```

```python
# КОД РИСУЕТ ГРАФИК ОТРЕЗКА ПРЯМОЙ ЛИНИИ В ПОЛЯРНЫХ КООРДИНАТАХ

R2 = np.linspace(1, 8, 10)
a2 = np.pi/2
b2 = R2 * np.cos(a2) + 50
plt.polar(b2, R2, 'green')
plt.xlabel('Line segment')
```
