```python
%matplotlib inline
import numpy as np

#!pip install termcolor

from termcolor import colored, cprint

for i in range(0, 20):
    a = int(input("Делайте ваши ставки, господа: "))
    cprint("Ставки сделаны, ставок больше нет", color='green')
    num = int(np.random.uniform(0, 36))
    red = [1, 3, 5, 7, 9, 12, 14, 16, 18, 19, 21, 23, 25, 27, 30, 32, 34, 36]
    black = [2, 4, 6, 8, 10, 11, 13, 15, 17, 20, 22, 24, 26, 28, 29, 31, 33, 35]
    if num == a:
        if num == 0:
            cprint("-------------You are the champion, my friend!!!----------", on_color='on_red', color='grey')
            cprint(" ZERO  GREEN ", on_color='on_green')
        elif num in red:
            cprint("-------------You are the champion, my friend!!!----------", on_color='on_red', color='grey')
            cprint(f" {num} RED ", on_color='on_red')
        elif num in black:
            cprint("-------------You are the champion, my friend!!!----------", on_color='on_red', color='grey')
            cprint(f" {num} BLACK ", on_color='on_grey')
        else:
            print('something wrong')
    else:
        if num == 0:
            cprint(" ZERO  GREEN ", on_color='on_green')
        elif num in red:
            cprint(f"{num} RED", color='grey', on_color='on_red')
        elif num in black:
            cprint(f"{num} BLACK", color='white', on_color='on_grey')
        else:
            print('something wrong')
            
```
